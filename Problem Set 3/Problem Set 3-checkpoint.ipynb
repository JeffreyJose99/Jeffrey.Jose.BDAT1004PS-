Assign it to a variable called data and replace the first 3 columns by a proper

```python
import pandas as pd
import numpy as np
import datetime
data=pd.read_csv("wind.txt",sep="\s+",parse_dates=[[0,1,2]])
data
```


Year 2061? Do we really have data from this year? Create a function to fix it
and apply it.

```python
def fixtime(x):
    year=x.year-100 if x.year > 1979 else x.year
    return datetime.date(year,x.month,x.day)
data['Yr_Mo_Dy']=data['Yr_Mo_Dy'].apply(fixtime)
data
```

Set the right dates as the index. Pay attention at the data type, it should be
datetime64[ns].


```python
data['Yr_Mo_Dy']=pd.to_datetime(data['Yr_Mo_Dy'])
data=data.set_index('Yr_Mo_Dy')
data
```

Compute how many values are missing for each location over the entire
record.They should be ignored in all calculations below

```python
data.isnull().sum()
```

Compute how many non-missing values there are in total.


```python
data.notnull().values.ravel().sum()
```

Create a DataFrame called loc_stats and calculate the min, max and mean
windspeeds and standard deviations of the windspeeds at each location over all the
days
A different set of numbers for each location

```python
loc_stats=pd.DataFrame()
loc_stats['min']=data.min()
loc_stats['max']=data.max()
loc_stats['mean']=data.mean()
loc_stats['std']=data.std()
loc_stats
```

Create a DataFrame called day_stats and calculate the min, max and mean
windspeed and standard deviations of the windspeeds across all the locations at each

```python
day_stats=pd.DataFrame()
day_stats['min']=data.min(axis=1)
day_stats['max']=data.max(axis=1)
day_stats['mean']=data.mean(axis=1)
day_stats['std']=data.std(axis=1)
day_stats
```

Find the average windspeed in January for each location.
Treat January 1961 and January 1962 both as January.

```python
data['date']=data.index
data['month']=data['date'].apply(lambda date:date.month)
data['year']=data['date'].apply(lambda date:date.year)
data['day']=data['date'].apply(lambda date:date.day)
jan_windspeeds=data.query('month==1')
jan_windspeeds
jan_windspeeds.loc[:,'RPT':'MAL'].mean()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
columns = ["Marriages_per_1000","Divorces_per_1000"]

x_data = data['Year'].values
fig ,z = plt.subplots()

for col in columns:
    z.plot(x_data,data[col].values)

z.set_title("Number of marriages and divorces per capita in U.S. between 1867 and 2014")
z.legend(columns)

plt.xlabel("Years")
plt.ylabel("Number of marriages & divorces per capita")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
chart = data[data["Year"].apply(lambda x : x in [1900,1950,2000])]

labels = chart['Year'].tolist()
marriages = chart['Marriages_per_1000'].tolist()
divorces = chart['Divorces_per_1000'].tolist()

x = np.arange(len(labels))  
width = 0.4

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, marriages, width, label='marriages')
bars2 = ax.bar(x + width/2, divorces, width, label='divorces')

ax.set_ylabel('Marriages & Divorces per capita')
ax.set_xlabel('Year')
ax.set_title('Number of marriages and divorces percapita in the U.S. in 1900, 1950 and 2000.')
ax.legend()

plt.xticks(x, labels)

plt.show()
```
